# 色散和耗散分析

取Fourier展开的一个频率：
$$
u ( x , t ) = A ( t ) e ^ { i \omega x }
$$
精确导数：
$$
\frac { \partial u } { \partial x } ( x ) = i \omega A ( t ) e ^ { i \omega x } = \frac { i k } { h } u ( x , t )
$$
其中$0<k=\omega h\leq \pi$

近似导数：
$$
\frac { 1 } { h } \sum _ { j = - N } ^ { N } b _ { j } u ( x + j h ) = \frac { i k ^ { \prime } } { h } u ( x , t )
$$
考虑初值：$u ( x , 0 ) = A ( 0 ) e ^ { \frac { i k } { h } x }$，精确解：
$$
u ( x , t ) = A ( 0 ) e ^ { \frac { i k } { h } ( x - a t ) }
$$
近似解：
$$
u ( x , t ) = A ( 0 ) e ^ { \frac { 1 } { h } \operatorname { Im } \left( k ^ { \prime } \right) a t } e ^ { \frac { i k } { h } \left( x - \frac { \operatorname { Re } \left( k ^ { \prime } \right) } { k } a t \right) }
$$
修正波数：
$$
\operatorname { Re } \left( k ^ { \prime } \right) = \sum _ { j = - N } ^ { N } b _ { j } \sin ( j k )
$$

$$
\quad \operatorname { Im } \left( k ^ { \prime } \right) = - \sum _ { j = - N } ^ { N } b _ { j } \cos ( j k ) _ { \circ }
$$

对一阶迎风格式：$b_{-1}=-1$, $b_0=1$
$$
\operatorname { Re } \left( k ^ { \prime } \right) =-\sin({-k})=\sin k
$$

$$
\quad \operatorname { Im } \left( k ^ { \prime } \right) = \cos{k}-1
$$

对二阶中心格式，代入$a=1​$有：
$$
a \frac { \partial u } { \partial x } =a \frac { 0.3 u _ { j + 1 }+ 0.4u_j -0.7 u _ { j - 1 } } { 2 \Delta x }
$$
$b_{-1}=-0.7$，$b_0=0.4$, $b_{1}=0.3$
$$
\operatorname { Re } \left( k ^ { \prime } \right) = -0.7\sin({-k})+0.3\sin({k})=\sin{k}
$$

$$
\quad \operatorname { Im } \left( k ^ { \prime } \right) =0.7\cos({-k})-0.4-0.3\cos({k})=0.4\cos{k}-0.4
$$
