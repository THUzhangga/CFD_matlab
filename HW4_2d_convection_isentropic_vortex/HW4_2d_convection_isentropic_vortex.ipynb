# 等熵涡问题

试用有限差分方法求解下列问题：

 

等熵涡问题，详见

条件：

方程：二维Euler方程（直角坐标系）

网格：正方形均匀网格

边界条件：**周期性边界条件**

 

可以使用中心型差分格式，如Lax-Wendroff， MacComack， Jameson等,或者二阶、三阶迎风格式求解。也可以用自己设计的计算格式求解。

 

要求：改变网格数量计算，分析网格数对计算结果精度的影响，求出格式误差的下降速率。比较至少两种差分格式的计算效果。（本例有精确解，可用误差随网格的变化估计整体误差的收敛阶次；此外，当网格加密时，时间步长也要按比例减小，这样就可以保证当时间精度不低于空间精度时，收敛阶次主要空间精度决定。）

Two-dimensional conservation law:
$$
\frac { \partial u } { \partial t } + \frac { \partial F ( u ) } { \partial x } + \frac { \partial G ( u ) } { \partial y } = 0
$$
Two-dimensional vortex evolution problem for the Euler equations. We consider the compressible Euler equations of gas dynamics:


$$
U _ { t } + F ( U ) _ { x } + G ( U ) _ { y } = 0
$$
其中：
$$
\begin{aligned} U & = ( \rho , \rho u , \rho v ,\rho E ) \\ 
F ( U ) & = \left( \rho u , \rho u ^ { 2 } + p , \rho u v , u (\rho E + p ) \right) \\ 
G ( U ) & = \left( \rho v , \rho u v , \rho v ^ { 2 } + p , v (\rho E + p ) \right) \end{aligned}
$$
$\rho​$是密度，$(u,v)​$是速度，$E​$是总能，$p​$是压力，且：
$$
E = \frac { p } { \gamma - 1 } + \frac { 1 } { 2 } \rho \left( u ^ { 2 } + v ^ { 2 } \right)
$$
$\gamma$是气体绝热指数，为1.4。

对于常规流动，取$\rho​$为1，$p​$=1，$(u,v)=(1,1)​$。对流体施加等熵涡（对$(u,v)​$和温度$T=p/\rho​$施加扰动，对熵$S=p/\rho^\gamma​$无扰动）：
$$
\begin{aligned} ( \delta u , \delta v ) & = \frac { \epsilon } { 2 \pi } e ^ { 0.5 \left( 1 - r ^ { 2 } \right) } ( - \overline { y } , \overline { x } ) \\ \delta T & = - \frac { ( \gamma - 1 ) \epsilon ^ { 2 } } { 8 \gamma \pi ^ { 2 } } e ^ { 1 - r ^ { 2 } } , \quad \delta S = 0 \end{aligned}
$$
其中：$( \overline { x } , \overline { y } ) = ( x - 5 , y - 5 ) , r ^ { 2 } = \overline { x } ^ { 2 } + \overline { y } ^ { 2 }$，涡强度$\epsilon=5​$

## 求解

考虑Euler方程的拟线性形式：
$$
\frac { \partial ( U ) } { \partial t } +  { \boldsymbol { A } } \frac { \partial } { \partial x } ( U ) +  { \boldsymbol { B } } \frac { \partial } { \partial y } ( U ) = 0
$$

系数矩阵$\boldsymbol{A}$和$\boldsymbol{B}$的推导在附录中完成。参照附录A，计算二维Euler方程在曲线坐标系中Jacobi矩阵的左右特征向量。

### 二阶迎风格式SCM方法：

把$ { \boldsymbol{A} }$和$ { \boldsymbol{B} }$分解为正负两个部分：
$$
\begin{array} { l } {  { \boldsymbol{A} } =  { \boldsymbol{A} } ^ { + } +  { \boldsymbol{A} } ^ { - } } \\ {  { \boldsymbol{B} } =  { \boldsymbol { B } } ^ { + } +  { \boldsymbol { B } } ^ { - } } \end{array}
$$

按照附录A所示方法计算左右特征向量。拟线性方程可以写为：
$$
\frac { \partial (  { \boldsymbol { U } } ) } { \partial t } +  { \boldsymbol { A } } ^ { + } \frac { \partial } { \partial x } (  { \boldsymbol { U } } ) +  { \boldsymbol { A } }^{-} \frac { \partial } { \partial x } (  { \boldsymbol { U } } ) +  { \boldsymbol { B } } ^ { + } \frac { \partial } { \partial y } (  { \boldsymbol { U } } ) +  { \boldsymbol { B } }^{-} \frac { \partial } { \partial y } (  { \boldsymbol { U } } ) = 0
$$
对${ \boldsymbol{A} } ^ { + }$和${ \boldsymbol{B} } ^ { + }$采用后差离散，对${ \boldsymbol{A} } ^ { - }$和${ \boldsymbol{B} } ^ { - }$采用前差离散。
$$
\begin{aligned}
U _ { i, j } ^ { n + 1 } &= U _ { i, j } ^ { n } \\
& - \frac { \Delta t } { 2\Delta x } \left[ A _ { i, j } ^ { - } \left( - U _ { i, j + 2 } ^ { n }  +  4 U _ { i, j + 1 } ^ { n } -3 U _ { i, j } ^ { n } \right) + A _ { i, j } ^ { + } \left( 3 U _ { i, j } ^ { n } - 4 U _ { i, j - 1 } ^ { n } +  U _ { i, j - 2 } ^ { n } \right) \right] \\
& - \frac { \Delta t } { 2\Delta y } \left[ B _ { i, j } ^ { - } \left( - U _ { i+2, j } ^ { n }  +  4 U _ { i+1, j } ^ { n } -3 U _ { i, j } ^ { n } \right) + B _ { i, j } ^ { + } \left( 3 U _ { i, j } ^ { n } - 4 U _ { i - 1, j} ^ { n } +  U _ { i - 2, j} ^ { n } \right) \right]
\end{aligned}
$$

### Lax–Wendroff格式

由课本P147公式6.1.9，直角坐标系下Lax-Wendroff格式的形式为：
$$
\begin{aligned} 
U _ { i , j } ^ { n + 1 } &= U _ { i , j } ^ { n } - \left( \frac { \boldsymbol { F } _ { i + 1 , j } ^ { n } - \boldsymbol { F } _ { i - 1 , j } ^ { n } } { 2 \Delta x } + \frac { G _ { i , j + 1 } ^ { n } - G _ { i , j - 1 } ^ { n } } { 2 \Delta y } \right) \Delta t +  \\ 
&\frac { ( \Delta t ) ^ { 2 } } { 2 \Delta x } \left[\boldsymbol { A } _ { i + 1 / 2 , j } ^ { n } \left( \frac { \boldsymbol { F } _ { i + 1 , j } ^ { n } - \boldsymbol { F } _ { i , j } ^ { n } } { \Delta x } + \frac { G _ { i + 1 / 2 , j + 1 / 2 } - \boldsymbol { G } _ { i + 1 / 2 , j - 1 / 2 } ^ { n } } { \Delta y } \right) +\\
A _ { i - 1 / 2 , j } ^ { n } \left( \frac { \boldsymbol { F } _ { i , j } ^ { n } - \boldsymbol { F } _ { i - 1 , j } ^ { n } } { \Delta x } + \frac { \boldsymbol { G } _ { i - 1 / 2 , j + 1 / 2 } - \boldsymbol { G } _ { i - 1 / 2 , j - 1 / 2 } ^ { n } } { \Delta y } \right)
\right]+\\
&\frac { ( \Delta t ) ^ { 2 } } {2 \Delta y } \left[  { \boldsymbol { B } } _ { i , j + 1 / 2 } ^ { n } \left( \frac {  { \boldsymbol { F } } _ { i + 1 / 2 , j + 1 / 2 } ^ { n } -  { \boldsymbol { F } } _ { i - 1 / 2 , j + 1 / 2 } ^ { n } } { \Delta x } + \frac {  { \boldsymbol { G } } _ { i , j + 1 } ^ { n } -  { \boldsymbol { G } } _ { i , j } ^ { n } } { \Delta y } \right) -\\
{ \boldsymbol { B } } _ { i , j - 1 / 2 } ^ { n } \left( \frac {  { \boldsymbol { F } } _ { i + 1 / 2,2 , j / 2 } ^ { n } -  { \boldsymbol { F } } _ { i - 1 / 2 , j - 1 / 2 } ^ { n } } { \Delta x } + \frac {  { \boldsymbol { G } } _ { i , j } ^ { n } -  { \boldsymbol { G } } _ { i , j - 1 } ^ { n } } { \Delta y } \right) \right]
\end{aligned}
$$
其中，涉及到角点$( i \pm 1 / 2 , j \pm 1 / 2 )$的通量由周围整节点通量函数的算数平均得到：
$$
 { \boldsymbol { F } } _ { i + 1 / 2 , j + 1 / 2 } ^ { n } = \frac { 1 } { 4 } \left(  { \boldsymbol { F } } _ { i , j } ^ { n } +  { \boldsymbol { F } } _ { i + 1 , j } ^ { n } +  { \boldsymbol { F } } _ { i , j + 1 } ^ { n } +  { \boldsymbol { F } } _ { i + 1 , j + 1 } ^ { n } \right)
$$
半节点的Jacobi矩阵采用整节点Jacobi的算数平均得到：
$$
{ A } _ { i + 1 / 2 , j } ^ { n } = \frac { 1 } { 2 } \left(  { A } _ { i , j } ^ { n } +  { A } _ { i + 1 , j } ^ { n } \right)
$$
